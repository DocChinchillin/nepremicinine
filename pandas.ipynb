```python
def readDeliInCene(year):
    tip = 'DELISTAVB'
    data_path = f'podatki/ETN_SLO_{year}_KPP_20240721/ETN_SLO_{year}_KPP_KPP_{tip}_20240721.csv'
    df = pd.read_csv(data_path)
    tip = 'POSLI'

    # cist useless al pa null
    df.drop(columns=["PRODANA_UPORABNA_POVRSINA_DELA_STAVBE",
    "NADSTROPJE_DELA_STAVBE",
    "STEVILO_ZUNANJIH_PARKIRNIH_MEST",
    "ATRIJ",
    "POVRSINA_ATRIJA",
    "OPOMBE_O_NEPREMICNINI",
    "PARCELNA_STEVILKA_ZA_GEOLOKACIJO",
    "INTERNA_OZNAKA_DELA_STAVBE",
    "EVIDENTIRANOST_DELA_STAVBE",
    "STEVILO_SOB",
    "POGODBENA_CENA_DELA_STAVBE",
    "STOPNJA_DDV_DELA_STAVBE"
    ], inplace=True)

    #kinda useless al pa null
    df.drop(columns=["PRODANA_POVRSINA_DELA_STAVBE"
    ], inplace=True)

    # posli
    posli_path = f'podatki/ETN_SLO_{year}_KPP_20240721/ETN_SLO_{year}_KPP_KPP_{tip}_20240721.csv'
    df_posli = pd.read_csv(posli_path)

    df_posli.drop(columns=[
        "DATUM_IZTEKA_LIZINGA",
        "DATUM_PRENEHANJA_LIZINGA",
        "LETO"
    ], inplace=True) 

    return pd.merge(left=df, right= df_posli, left_on='ID_POSLA', right_on='ID_POSLA')
```

```python
df = readDeliInCene(startYear)
for leto in range(startYear + 1, endYear + 1):
    df = pd.concat([df,readDeliInCene(leto)], axis=0)
```

```python
df.query("PRODANA_POVRSINA > 0",inplace=True)
```

```python
def whichRegija(obcina):
    for imeRegije,obcineRegije in regije.items():
        try:
            if obcina.upper() in obcineRegije:
                return imeRegije
        except:
            return "Brez"
    return "Brez"
df["REGIJA"] = df["OBCINA"].apply(whichRegija)
df["CENA_NA_M2"] = df["POGODBENA_CENA_ODSKODNINA"] / df["PRODANA_POVRSINA"]
```

```python
# Group by ID_POSLA and calculate unique counts for VRSTA_DELA_STAVBE and REGIJA
if True:
    grouped = df.groupby('ID_POSLA').agg({
        'VRSTA_DELA_STAVBE': pd.Series.nunique,
        'REGIJA': pd.Series.nunique
    })

    # 
    filtered_id_posla = grouped[(grouped['VRSTA_DELA_STAVBE'] <= 1) & (grouped['REGIJA'] <= 1)].index


    # filtriramo posle, ki so del več regij in imajo različne stavbe
    df = df[df['ID_POSLA'].isin(filtered_id_posla)]
```

```python
if True:
    # Identify ID_POSLA with duplicates
    duplicate_ids = df['ID_POSLA'].duplicated(keep=False)

    # Filter the DataFrame to remove duplicates
    df = df[~duplicate_ids]
```
